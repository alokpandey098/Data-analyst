- üé® 1.1 What is Matplotlib?

Matplotlib is the foundation of Python data visualization.
Everything in Seaborn builds on top of it.
Think of Matplotlib as a blank canvas.
You manually tell it what to draw:
  - X, Y data points
  - Colors, titles, labels, legends
It gives low-level control ‚Äî perfect for fine-tuning.

```python
from matplotlib import pyplot as plt
x = [11,2,33,40,5]
y = [2,4,6,8,10]

plt.plot(x,y,color='black',marker='*')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
plt.plot(x,y)
```

```python
plt.bar(x,y)
```

```python
# Line plot variation

x = [10,20,33,44,59]
y = [20,44,60,88,100]

plt.plot(x,y,color='red',linestyle='--',marker='+',linewidth=3)
plt.title('Cutomize line chart')
plt.grid(True)
plt.show()
```

- üß† Concepts Learned:

1. color, linestyle, marker, linewidth

2. plt.grid() adds readability

```python
# Multiple lines
plt.plot(x, y, label='y = 2x')
plt.plot(x, [i**2 for i in x], label='y = x¬≤')
plt.legend()
plt.show()
```

1. Measure of central tendency (Mean ,Median, Mode) .
  * Mean

Mean or Average calculation :- sum of all values in data / Total no. of values

```python
# Dataset creation with numpy
ar = np.array([2,1,5,7,3,4,9,1,11,8,4,2,0])
print(ar)
```

```python
# Find mean or average

# Without built-in function
avrg = np.sum(ar)/len(ar)
print('Average of array is :',avrg)
```

```python
# With Built-in function
avrg = np.mean(ar)
print('Average of array is :',avrg)
```

* Median

Median Calculation :- Sort the values, then pick one value from middle = Ans .

(i) If count of values is even :- Sum of two values from middle/2 =Ans

(ii) If count of values is odd :- Middle value = Ans

```python
med = np.median(data['fare'])
print('Median value of fare is :',med)
```

```python
# Ploting Median alongwith mean
sns.histplot(x=data['age'],bins=[ i for i in range(0,81,10)])
plt.plot([avg for i in range(0,300)],[i for i in range(0,300)],c='orange')
plt.plot([med for i in range(0,300)],[i for i in range(0,300)],c='red')
plt.show()
```

(ii) - Mean absolute division (MAD)

Check the data is how much spread from mean .

```python
# Creating dataset for MAD
sec_a = np.array([75,65,73,68,72,76])
sec_b = np.array([90,47,43,96,93,51])
num = np.array([1,2,3,4,5,6])

mean_a = np.mean(sec_a)
mean_b = np.mean(sec_b)

print('Mean of A :',mean_a)
print('Mean of B :',mean_b)
```

```python
# Now find MAD 

mad_a = np.sum(np.abs(sec_a-mean_a)/len(sec_a))
mad_b = np.sum(np.abs(sec_b-mean_b)/len(sec_b))

print('MAD of sec_a :',mad_a)
print('MAD of sec_b :',mad_b)
```

```python
# Plot Figure
plt.figure(figsize=(10,3))
plt.scatter(sec_a,num,label='Sec-A')
plt.scatter(sec_b,num,label='Sec-B',color='red')
plt.plot([70,70,70,70,70,70],num,c='black',label='Mean')
plt.legend()
plt.plot()
```

less spread = perfect for algorithm

3. Percentage and Percentiles and Quartiles

- Percentage :- Percentage tells how much out of 100. Useful in data cleaning to find missing/invalid data ratio. Always divide by original value, not new value.

```python
data = {
    "score": [45, 55, 60, 65, 70, 75, 80, 85, 90, 95],
    "gender": ["M", "F", "M", "F", "M", "M", "F", "F", "M", "F"]
}

df = pd.DataFrame(data)
```

```python
female_percentage = (df['gender'] == 'F').mean() * 100
print(female_percentage)
```

```python
# What percentage of students scored above 70?
std = (df['score'] > 70).mean() * 100
print(std)
```

- Percentile :- 50th percentile means 50% of data lies below this value. Percentiles show relative position in a dataset (very common in salary & performance analysis).

```python
# Find the 50th and 90th percentile of scores.
df['score'].quantile([0.5, 0.9])
```

```python
#Find students scoring above the 75th percentile.
percentile_75 = df['score'].quantile(0.75)
df[df['score'] > percentile_75]
```

- Quartile :- Q1  25% data below, Q2  50% data below, Q3  75% data below .
  Used heavily in box plots & outlier detection.

```python
df['score'].quantile([0.25, 0.5, 0.75])
```

```python
# Find the Interquartile Range(IQR)

Q1 = df['score'].quantile(0.25)
Q3 = df['score'].quantile(0.75)
IQR = Q3 - Q1
print("IQR value :",IQR)
```

* Explanation : IQR is used for outlier detection.

5. Probability

```python
# Q- Estimate the probability that a randomly selected student is male.

prob = (df['gender'] == 'M').mean()
print(prob)
```

* Explanation - Probability  relative frequency in data.

```python
# Q - What is the probability that a student scored between 60 and 80?

marks_prob = ((df['score'] >= 60) & (df['score'] <= 80)).mean()
print('probability that a student scored between 60 and 80 :',marks_prob)
```

* Explanation - Logical conditions + mean = probability.

```python
# Q - Find the skewness of the score distribution.
skeww = df['score'].skew()
print(skeww)
```

* Explanation : ( > 0  Right skewed ) ,( < 0  Left skewed )
, (  0  Symmetric )

```python
# Q - Find the kurtosis of the scores.

kurtosiss = df['score'].kurtosis()
print(kurtosiss)
```

* Explanation : High  heavy tails (outliers) , Low  flat distribution

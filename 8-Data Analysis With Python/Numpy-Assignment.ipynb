### Assignment 4: Statistical Operations

1. Create a NumPy array of shape (5, 5) filled with random integers. Compute the mean, median, standard deviation, and variance of the array.
2. Create a NumPy array of shape (3, 3) with values from 1 to 9. Normalize the array (i.e., scale the values to have a mean of 0 and a standard deviation of 1).

```python
# Question 1.

arry = np.random.randint(1,40,size=(5,5))
print('Original Array :')
print(arry)

# mean, median, standard deviation, and variance

arr_mean = np.mean(arry)
arr_med = np.median(arry)
arr_std = np.std(arry)
arr_var = np.var(arry)

print("\nMean:", arr_mean)
print("Median:", arr_med)
print("Standard Deviation:", arr_std)
print("Variance:", arr_var)
```

```python
import numpy as np

# Create 3x3 array with values 1 to 9
arr = np.arange(1, 10).reshape(3, 3)

print("Original Array:")
print(arr)

# Compute mean and std
mean_val = np.mean(arr)
std_val = np.std(arr)

# Normalize the array
normalized_arr = (arr - mean_val) / std_val

print("\nNormalized Array:")
print(normalized_arr)

print("\nMean after normalization:", normalized_arr.mean())
print("Std after normalization:", normalized_arr.std())
```

### Assignment 5: Broadcasting

1. Create a NumPy array of shape (3, 3) filled with random integers. Add a 1D array of shape (3,) to each row of the 2D array using broadcasting.
2. Create a NumPy array of shape (4, 4) filled with random integers. Subtract a 1D array of shape (4,) from each column of the 2D array using broadcasting.

### Assignment 6: Linear Algebra

1. Create a NumPy array of shape (3, 3) representing a matrix. Compute its determinant, inverse, and eigenvalues.
2. Create two NumPy arrays of shape (2, 3) and (3, 2). Perform matrix multiplication on these arrays.

```python
# Question 1.
import numpy as np

# Create a 3x3 matrix (use floats to avoid integer division surprises)
A = np.array([[4.0, 7.0, 2.0],
              [3.0, 5.0, 1.0],
              [2.0, 4.0, 3.0]])

np.set_printoptions(precision=6, suppress=True)  # nicer printing

print("Matrix A:")
print(A)

# Determinant
detA = np.linalg.det(A)
print("\nDeterminant of A:", detA)

# Inverse (only if determinant is not zero / not too close to zero)
if abs(detA) < 1e-12:
    print("\nMatrix is singular (determinant ~ 0). Inverse does not exist.")
else:
    A_inv = np.linalg.inv(A)
    print("\nInverse of A:")
    print(A_inv)

# Eigenvalues and eigenvectors
eigvals, eigvecs = np.linalg.eig(A)
print("\nEigenvalues of A:")
print(eigvals)
print("\nEigenvectors (each column is an eigenvector):")
print(eigvecs)
```

### Assignment 9: Structured Arrays

1. Create a structured array with fields 'name' (string), 'age' (integer), and 'weight' (float). Add some data and sort the array by age.
2. Create a structured array with fields 'x' and 'y' (both integers). Add some data and compute the Euclidean distance between each pair of points.

```python
# Question 1.
# Create structured array data type
dtype = [('name', 'U10'), ('age', 'i4'), ('weight', 'f4')]

# Add data
data = np.array([
    ('Alok', 24, 68.5),
    ('Ravi', 19, 55.2),
    ('Sneha', 22, 60.3),
    ('Mohan', 28, 72.0)
], dtype=dtype)

print("Original Array:")
print(data)

# Sort by age
sorted_data = np.sort(data, order='age')

print("\nSorted by Age:")
print(sorted_data)
```

```python
# Question 2.
# Create structured array for 2D points
dtype = [('x', 'i4'), ('y', 'i4')]

points = np.array([
    (1, 2),
    (3, 4),
    (6, 8),
    (2, 1)
], dtype=dtype)

print("Points:")
print(points)

# Extract x and y as arrays
x = points['x']
y = points['y']

# Compute Euclidean distance from origin (0,0)
distances = np.sqrt(x**2 + y**2)

print("\nEuclidean Distance of Each Point from Origin:")
print(distances)
```

```python
from scipy.spatial.distance import pdist, squareform

coords = np.column_stack((points['x'], points['y']))

pairwise_distances = squareform(pdist(coords))

print("\nPairwise Euclidean Distance Matrix:")
print(pairwise_distances)
```

### Assignment 10: Masked Arrays

1. Create a masked array of shape (4, 4) with random integers and mask the elements greater than 10. Compute the sum of the unmasked elements.
2. Create a masked array of shape (3, 3) with random integers and mask the diagonal elements. Replace the masked elements with the mean of the unmasked elements.

```python
# Question 1.
import numpy.ma as ma

# Create random array
arr = np.random.randint(1, 20, size=(4, 4))
print("Original Array:\n", arr)

# Mask values greater than 10
masked_arr = ma.masked_greater(arr, 10)
print("\nMasked Array:\n", masked_arr)

# Sum of unmasked values
sum_unmasked = masked_arr.sum()
print("\nSum of Unmasked Elements:", sum_unmasked)
```

```python
# Question 2.
# Create random 3x3 array
arr2 = np.random.randint(1, 20, size=(3, 3))
print("Original Array:\n", arr2)

# Mask diagonal elements
masked_arr2 = ma.array(arr2, mask=np.eye(3, dtype=bool))
print("\nMasked Array:\n", masked_arr2)

# Compute mean of unmasked elements
mean_unmasked = masked_arr2.mean()
print("\nMean of Unmasked Elements:", mean_unmasked)

# Fill masked elements with the mean
filled_arr = masked_arr2.filled(mean_unmasked)
print("\nArray After Replacing Masked Diagonal:\n", filled_arr)
```

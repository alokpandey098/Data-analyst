- üé® 1.1 What is Matplotlib?

Matplotlib is the foundation of Python data visualization.
Everything in Seaborn builds on top of it.
Think of Matplotlib as a blank canvas.
You manually tell it what to draw:
  - X, Y data points
  - Colors, titles, labels, legends
It gives low-level control ‚Äî perfect for fine-tuning.

```python
from matplotlib import pyplot as plt
x = [11,2,33,40,5]
y = [2,4,6,8,10]

plt.plot(x,y,color='black',marker='*')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# Line plot variation

x = [10,20,33,44,59]
y = [20,44,60,88,100]

plt.plot(x,y,color='red',linestyle='--',marker='+',linewidth=3)
plt.title('Cutomize line chart')
plt.grid(True)
plt.show()
```

```python
# Histogram
import numpy as np
data = np.random.randn(1000)
plt.hist(data, bins=20, color='orange', edgecolor='black')
plt.title("Histogram Example")
plt.show()
```

```python
# Pie chart
sizes = [20, 30, 25, 25]
labels = ['Apple', 'Banana', 'Grapes', 'Mango']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart Example")
plt.show()
```

```python
# Boxplot

data = [20, 21, 23, 25, 30, 35, 40]
plt.boxplot(data)
plt.title("Box Plot Example")
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1, y2, y3 = np.sin(x), np.cos(x), np.tan(x)

fig, axs = plt.subplots(2, 2, figsize=(10, 6))

axs[0, 0].plot(x, y1, color='blue', label='sin(x)')
axs[0, 1].plot(x, y2, color='red', label='cos(x)')
axs[1, 0].plot(x, y3, color='green', label='tan(x)')

# Titles and labels
for ax in axs.flat:
    ax.legend()
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')

plt.tight_layout()
plt.show()
```

- üß† Tip: Use figsize, tight_layout(), and sharex/sharey for a clean layout.

```python
fig, ax1 = plt.subplots()

x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [1, 3, 5, 7, 9]

ax1.plot(x, y1, color='blue', label='Revenue')
ax2 = ax1.twinx()
ax2.plot(x, y2, color='orange', label='Growth')

ax1.set_ylabel('Revenue')
ax2.set_ylabel('Growth')
ax1.set_xlabel('Year')

plt.title('Revenue vs Growth')
plt.show()
```

```python
# Adding Text, Annotations & Arrows

plt.plot(x, y1)
plt.title("Annotated Sine Wave")
plt.annotate('Peak Point', xy=(1.5, 1), xytext=(2, 1.2),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

```python
# Custom Color Maps & Themes
import matplotlib.cm as cm

plt.scatter(x, y1, c=y1, cmap='plasma', s=80)
plt.colorbar(label='Intensity')
plt.title('Custom Colormap Example')
plt.show()
```

```python
# 3D Plotting
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')
plt.title('3D Surface Plot')
plt.show()
```

```python
# Interactivity with Matplotlib Widgets (Optional Advanced)

from matplotlib.widgets import Slider

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
fig, ax = plt.subplots()
line, = ax.plot(x, y)
plt.subplots_adjust(bottom=0.25)

ax_slider = plt.axes([0.25, 0.1, 0.65, 0.03])
slider = Slider(ax_slider, 'Freq', 0.1, 10, valinit=1)

def update(val):
    line.set_ydata(np.sin(x * val))
    fig.canvas.draw_idle()

slider.on_changed(update)
plt.show()
```

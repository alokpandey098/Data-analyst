```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

1. Measure of central tendency (Mean ,Median, Mode) .
  * Mean

Mean or Average calculation :- sum of all values in data / Tatal no. of values

```python
# Dataset creation with numpy
ar = np.array([2,1,5,7,3,4,9,1,11,8,4,2,0])
print(ar)
```

```python
# Find mean or average

# Without built-in function
avrg = np.sum(ar)/len(ar)
print('Average of array is :',avrg)
```

```python
# With Built-in function
avrg = np.mean(ar)
print('Average of array is :',avrg)
```

```python
# Load dataset
data = sns.load_dataset('titanic')
```

```python
data.head(3)
```

```python
data.isnull().sum()
```

```python
# Fill Nan Values

data['age'].fillna(data['age'].mean(),inplace=True)
```

```python
avg = np.mean(data['age'])
print('Average age of passenger is :',round(avg,2),'Year')
```

```python
# Ploting mean with actual value 
sns.histplot(x=data['age'],bins=[ i for i in range(0,81,10)])
plt.plot([avg for i in range(0,300)],[i for i in range(0,300)],c='black')
plt.show()
```

* Median

Median Calculation :- Sort the values, then pick one value from middle = Ans .

(i) If count of values is even :- Sum of two values from middle/2 =Ans

(ii) If count of values is odd :- Middle value = Ans

```python
med = np.median(data['fare'])
print('Median value of fare is :',med)
```

```python
# Ploting Median alongwith mean
sns.histplot(x=data['age'],bins=[ i for i in range(0,81,10)])
plt.plot([avg for i in range(0,300)],[i for i in range(0,300)],c='orange')
plt.plot([med for i in range(0,300)],[i for i in range(0,300)],c='red')
plt.show()
```

* Mode (Used for Categorical or discrete value)

Mode calculation :- Most frequent value in dataset = Ans

```python
data.head()
```

```python
# Mode 
data['deck'].mode()[0]
```

```python
data['sex'].mode()[0]
```

```python
data['sex'].value_counts()
```

```python
mod = data['fare'].mode()[0]
print('Mode of fare is :',mod)
```

```python
# Ploting Median alongwith mean
sns.histplot(x=data['age'],bins=[ i for i in range(0,81,10)])
plt.plot([avg for i in range(0,300)],[i for i in range(0,300)],c='orange',label='Mean')
plt.plot([med for i in range(0,300)],[i for i in range(0,300)],c='red',label='Median')
plt.plot([mod for i in range(0,300)],[i for i in range(0,300)],c='black',label='Mode')
plt.legend()
plt.show()
```

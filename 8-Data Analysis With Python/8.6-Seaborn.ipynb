#### Seaborn = The Artist ðŸŽ¨

- Seaborn sits on top of Matplotlib â€” itâ€™s smarter and prettier.
It automatically handles:

    - Styling
    - Colors
    - Statistical relationships

```python
import seaborn as sns
import matplotlib.pyplot as plt

tip = sns.load_dataset('tips')
sns.scatterplot(data=tip,x='total_bill',y='tip',hue='sex')
plt.show()
```

- ðŸŽ¯ Matplotlib â†’ low-level, customizable
- ðŸŽ¯ Seaborn â†’ high-level, fast, beautiful

```python
list(tip.columns)
```

```python
sns.lineplot(data=tip,x='time',y='smoker')
```

```python
sns.barplot(data=tip,x='smoker',y='time',hue='sex')
```

```python
sns.barplot(x='day', y='total_bill', data=tip, hue='sex')
plt.title("Seaborn Bar Plot with Hue")
plt.show()
```

```python
# Box plot

sns.boxplot(x='day', y='total_bill', data=tip, hue='sex')
plt.title("Box Plot with Hue")
plt.show()
```

```python
# Violin plot 

sns.violinplot(x='day', y='total_bill', data=tip, hue='sex', split=True)
plt.title("Violin Plot Example")
plt.show()
```

```python
# Strip plot 

sns.stripplot(x='day', y='total_bill', data=tip, jitter=True)
plt.title("Strip Plot Example")
plt.show()
```

```python
# Histplot

sns.histplot(data=tip, x='total_bill', bins=20, kde=True,color='orange')
plt.title("Histogram + KDE")
plt.show()
```

```python
# Joint plot
sns.jointplot(x='total_bill', y='tip', data=tip, kind='scatter')
plt.show()
```

```python
# Pair plot 

sns.pairplot(tip, hue='sex')
plt.show()
```

```python
# Correlation Heatmap

iris = sns.load_dataset("iris")
corr = iris.corr(numeric_only=True) # For avoiding string value error
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Count Plot

sns.countplot(x='day', data=tip, hue='sex')
plt.title("Count of Days by Gender")
plt.show()
```

```python
# Swarmplot 

sns.swarmplot(x='day', y='total_bill', data=tip)
plt.title("Swarm Plot Example")
plt.show()
```

```python
# PairPlot (Relationship Among All Variables)

sns.pairplot(iris, hue="species", palette="Set2")
plt.show()
```

```python
# SwarmPlot & ViolinPlot
sns.violinplot(x="species", y="sepal_length", data=iris)
sns.swarmplot(x="species", y="sepal_length", data=iris, color='black', alpha=0.7)
plt.show()
```

```python
# FacetGrid (Multiple Charts by Category)

g = sns.FacetGrid(iris, col="species")
g.map(sns.histplot, "sepal_length", bins=10, color="skyblue")
plt.show()
```

```python
# Styling & Themes

sns.set_theme(style="whitegrid", palette="pastel")

sns.boxplot(x="species", y="petal_length", data=iris)
plt.title("Boxplot with Custom Theme")
plt.show()
```

```python
# Combine Matplotlib + Seaborn

sns.barplot(x="species", y="petal_length", data=iris)
plt.title("Customized Seaborn Plot")
plt.xlabel("Flower Species")
plt.ylabel("Petal Length (cm)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#  Advanced Color Palettes

sns.set_palette("rocket")
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.show()

# or manually
colors = sns.color_palette(["#ff9999", "#66b3ff", "#99ff99"])
sns.barplot(x="species", y="sepal_length", data=iris, palette=colors)
```

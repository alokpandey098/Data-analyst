### Assignment 1: Array Creation and Manipulation

1. Create a NumPy array of shape (5, 5) filled with random integers between 1 and 20. Replace all the elements in the third column with 1.
2. Create a NumPy array of shape (4, 4) with values from 1 to 16. Replace the diagonal elements with 0.

```python
import numpy as np
```

```python
# Queastion 1.

# Create a (5x5) array with random integers between 1 and 20
arr = np.random.randint(1, 21, size=(5, 5))

print("Original Array:")
print(arr)

# Replace all elements in the third column with 1
arr[:, 2] = 1     # column index 2 → third column

print("\nModified Array (3rd column replaced with 1):")
print(arr)
```

```python
# Question 2.

arr1 = np.arange(0,16).reshape(4,4)
print('Orignal Array :')
print(arr1)

# Replace diagonal elements with 0
np.fill_diagonal(arr1, 0)

print("\nModified Array (Diagonal elements replaced with 0):")
print(arr1)
```

### Assignment 2: Array Indexing and Slicing

1. Create a NumPy array of shape (6, 6) with values from 1 to 36. Extract the sub-array consisting of the 3rd to 5th rows and 2nd to 4th columns.
2. Create a NumPy array of shape (5, 5) with random integers. Extract the elements on the border.

```python
# Question 1.

arr2 = np.arange(1,37).reshape(6,6)
print('Array Before Exract ..')
print(arr2)

# Extract rows 3 to 5 → index 2 to 4
# Extract columns 2 to 4 → index 1 to 3

sub_arr2 = arr2[2:5 ,1:3]

print('\nSub array :  ')
print(sub_arr2)
```

```python
# Question 2.

rand_arr = np.random.randint(1,100,size=(5,5))
print('Original Array :')
print(rand_arr)

# rand_arr[ start_row : end_row ,start_col:end:col ]
top_row = rand_arr[0, :] 
last_row = rand_arr[-1,:]
left_col= rand_arr[1:-1,0]
right_col = rand_arr[1:-1,-1]

border_element = np.concatenate([top_row,right_col,last_row[::-1],left_col[::-1]])

print('\nBorder elements from original array :')
print(border_element)
```
